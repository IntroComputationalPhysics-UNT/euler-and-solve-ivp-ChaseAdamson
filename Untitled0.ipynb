```python
# euler solver for harmonic oscillator
def euler_sho(omega_0, theta_0, v_0, t_0, t_f, steps, implicit):
  """
  This function calculates the solution to the equation d^2/dt^2 theta + omega_0^2 sin(theta)=0 using Euler's method of numerical approximation.
  """
  dt = (t_f - t_0) / steps # calculates time interval dt
  theta = [theta_0] # list for theta values
  v = [v_0] # list for angular velocity values
  t = [t_0] # list for time values
  for _ in range(steps):
    theta_new = theta[-1] + v[-1] * dt # updates theta value
    if implicit: # boolean implicit(True) or explicit(False) method
      v_new = v[-1] - omega_0 ** 2 * math.sin(theta_new) * dt # update v : using implicit form
    else:
      v_new = v[-1] - omega_0 ** 2 * math.sin(theta[-1]) * dt # update v: using non implicit form
    t_new = t[-1] + dt # update t
    theta.append(theta_new) # appends theta to theta list
    v.append(v_new) # appends v to v list
    t.append(t_new) # append t to t list
  return t, theta, v

data=euler_sho(1, theta_0_1, ang_vel_0_1, 0, 100, 6000, True)
plt.figure(figsize=(10,2))
plt.xlim([0, 100]) # time
plt.ylim([- 5 * math.pi - 0.5, 5 * math.pi + 0.5]) # theta
plt.title("Figure 1: Pendulum 1")
plt.xlabel("Time")
plt.ylabel("Theta")
plt.plot(data[0], data[1])
```

```python
data=euler_sho(1, theta_0_2, ang_vel_0_2, 0, 100, 10**6, True)
plt.figure(figsize=(10,2))
plt.xlim([0, 100]) # time
plt.ylim([- 5 * math.pi - 0.5, 5 * math.pi + 0.5]) # theta
plt.title("Figure 2: Theta vs Time")
plt.xlabel("Time")
plt.ylabel("Theta")
plt.plot(data[0], data[1])

plt.figure(figsize=(10,2))
plt.xlim([-5*math.pi, 5*math.pi])
plt.ylim([-math.pi - 0.5, math.pi + 0.5]) # theta
plt.title("Figure 3: Phase Plot")
plt.xlabel("theta")
plt.ylabel("angular velocity")
plt.plot(data[1], data[2])
```

The implicit euler methods work better for longer time scales. The explicit method exhibits unstable behavior sooner than the implicit method. The number of steps needs to be extremely high for the second set of initial conditions to approach the expected behavior.

```python
def pendulum_EOMs(t, y):
  theta, ang_vel = y
  dtheta_dt=ang_vel
  dang_vel_dt=-np.sin(theta)
  return [dtheta_dt, dang_vel_dt]

t_span=(0, 100)
conditions_1=[theta_0_1, ang_vel_0_1]
conditions_2=[theta_0_2, ang_vel_0_2]

solution1 = solve_ivp(pendulum_EOMs, t_span, conditions_1, t_eval=np.linspace(0, 100, 500), method='RK45', max_step=1/120, atol=1e-9)
plt.plot(solution1.t, solution1.y[0], label="θ(t)")
plt.plot(solution1.t, solution1.y[1], label="ω(t)")
plt.title("Pendulum motion 1")
plt.xlabel("Time [s]")
plt.ylabel("θ, ω")
plt.legend()
plt.grid()
plt.show()

solution2 = solve_ivp(pendulum_EOMs, t_span, conditions_2, t_eval=np.linspace(0, 100, 500), method='RK45', max_step=1/120, atol=1e-9)
plt.plot(solution2.t, solution2.y[0], label="θ(t)")
plt.plot(solution2.t, solution2.y[1], label="ω(t)")
plt.title("Pendulum motion 2")
plt.xlabel("Time [s]")
plt.ylabel("θ, ω")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10,2))
plt.xlim([-2*math.pi, 2*math.pi])
plt.ylim([-math.pi - 0.5, math.pi + 0.5]) # theta
plt.title("Pendulum motion 2: Phase Plot")
plt.xlabel("theta")
plt.ylabel("angular velocity")
plt.plot(solution2.y[0], solution2.y[1])
```

The RK4 method does a far better job of preserving the stability of the simulation over time. The reduction of the time steps and error tolerances also makes the results more stable.

```python
# define our test euler method for a simple harmonic oscillator
def Euler_sho(omega_0, x_0, v_0, t_0, t_f, n_steps):
  """
  This function calculates the solution to x'' = - omega_0 ^2 x with Euler's method.
  """
  dt = (t_f - t_0) / n_steps # calculate dt
  x = [x_0] # initialized x list
  v = [v_0] # initialized v list
  t = [t_0] # initialized t list
  U=[1/2 * x_0**2]
  T=[1/2 * v_0**2]
  E=[1/2 * v_0**2 + 1/2 * x_0**2]
  for _ in range(n_steps):
    x_new = x[-1] + v[-1] * dt # update x
    v_new = v[-1] - omega_0 ** 2 * x_new * dt # update v : using an implicit form
    t_new = t[-1] + dt # update t
    U_new=1/2 * x_new**2 # update potential energy
    T_new=1/2 * v_new**2 # Update Kinetic energy
    E_new=U_new+T_new # update total energy
    x.append(x_new) # append x
    v.append(v_new) # append v
    t.append(t_new) # append t
    U.append(U_new) # append U
    T.append(T_new) # append T
    E.append(E_new) # append E
  return t, x, v, U, T, E

DATA=Euler_sho(1, 1, 0, 0, 100, 100*120)
plt.figure(figsize=(10,2))
plt.xlim([0, 100])
plt.ylim([-math.pi - 0.5, math.pi + 0.5])
plt.title("Spring Mass")
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.plot(DATA[0], DATA[1], label='Displacement')
plt.plot(DATA[0], DATA[2], label='Velocity')

plt.figure(figsize=(10,2))
plt.xlim([0, 100])
plt.ylim([-1/2, 1])
plt.title("Energy Plot")
plt.xlabel("Time")
plt.ylabel("Total energy")
plt.plot(DATA[0], DATA[3], label='Potential Energy')
plt.plot(DATA[0], DATA[4], label='Kinetic Energy')
plt.plot(DATA[0], DATA[5], label='Total Energy') # Sum of kinetic and potential energy
```

These energy contours show that energy is conserved through time.
